```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seconds = 1800   # 30 minutes data

time = np.arange(seconds)

hr = np.random.normal(80, 5, seconds)
spo2 = np.random.normal(98, 1, seconds)
bp_sys = np.random.normal(120, 5, seconds)
bp_dia = np.random.normal(80, 5, seconds)

motion = np.random.normal(0, 1, seconds)

# distress event
hr[600:700] += 40
spo2[600:700] -= 10

# motion artifact
spo2[1000:1020] -= 15
hr[1200:1210] += 30

df = pd.DataFrame({
    "time": time,
    "hr": hr,
    "spo2": spo2,
    "bp_sys": bp_sys,
    "bp_dia": bp_dia,
    "motion": motion
})

df.head()
```

```python
df.to_csv("vitals.csv", index=False)
```
